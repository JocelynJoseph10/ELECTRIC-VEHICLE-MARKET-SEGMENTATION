```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

### Loading the dataset

```python
df=pd.read_csv('RS_Session_254_AU_2384.C.i.csv')

df
```

###### The dataset was obtained from data.gov.in. The dataset ‘RS_Session_254_AU_2384.C.i.csv’ contains the total number of EVs registered under VAHAN4.0 based on their category and vehicle class. 

```python
print(df.shape)
print(df.info())
```

###### The dataset contains 27 rows and 4 columns. 

```python
rows_to_delete = [14,26]

# Delete the specified rows
df.drop(rows_to_delete, inplace=True)

# Reset the DataFrame index
df.reset_index(drop=True, inplace=True)
```

```python
df['Category'].unique()
```

```python
print(df['Vehicle Class'].unique())
print("\nThere are in total",df['Vehicle Class'].nunique()," vehicle classes")
```

#### DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")

# Create a bar plot to visualize the total registrations by category
plt.figure(figsize=(12, 6))
sns.barplot(x="Category", y="Total Registered on VAHAN 4.0", data=df)
plt.title('Total Vehicle Registrations by Category')
plt.xlabel('Category')
plt.ylabel('Total Registrations')
plt.xticks(rotation=45)
plt.show()

```

```python
# Data from the provided dataset
categories = ['Transport Vehicles', 'Non-Transport Vehicles']
total_registered = [550491, 169969]

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(total_registered, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vehicle Registrations')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

```python
# Create a bar plot to visualize the total registrations by vehicle class
plt.figure(figsize=(12, 8))
sns.barplot(x="Total Registered on VAHAN 4.0", y="Vehicle Class", data=df)
plt.title('Total Vehicle Registrations by Vehicle Class')
plt.xlabel('Total Registrations')
plt.ylabel('Vehicle Class')
plt.show()
```
